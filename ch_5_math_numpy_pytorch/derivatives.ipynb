```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import sympy.plotting.plot as symplot
from IPython.display import display
```

```python
x = sym.symbols('x')

fx = 2 * x**2
df = sym.diff(fx, x)
print(fx)
print(df)
```

```python
# plotting f and df

symplot(fx, (x, -4, 4), title='The function')
plt.show()

symplot(df, (x, -4, 4), title='Derivative')
plt.show()
```

```python
x = sym.symbols('x')
fx = sym.exp(x)
df = sym.diff(fx, x)
print(fx)
print(df)
```

```python
# we'll do the same for relu and sigmoid functions

relu = sym.Max(0, x)
sigmoid = 1 / (1 + sym.exp(-x))

# graph the functions
p = symplot(relu, (x, -4, 4), label='ReLU', show=False, line_color='blue')
p.extend(symplot(sigmoid, (x, -4, 4), label='Sigmoid', show=False, line_color='red'))
p.legend = True
p.title = 'The functions'
p.show()

# grpah the derivatives
p = symplot(sym.diff(relu), (x, -4, 4), label='df(ReLU)', show=False, line_color='blue')
p.extend(
    symplot(sym.diff(sigmoid), (x, -4, 4), label='df(Sigmoid)', show=False, line_color='red')
)
p.legend = True
p.title = 'The derivatives'
p.show()
```

```python
x = sym.symbols('x')

# functions
fx = 2*x**2
gx = 4*x**3 -3*x**4

# individual derivatives
df = sym.diff(fx)
dg = sym.diff(gx)

# apply the product rule "manually"
manual = df*gx + fx*dg
thewrongway = df*dg

sympy_derivative = sym.diff(fx * gx)

print('The functions:')
display(fx)
display(gx)
print(' ')

print('The derivaties of functions:')
display(df)
display(dg)
print(' ')

print('Manual product rule:')
display(manual)
print(' ')

print('Via sympy: ')
display(sympy_derivative)
print(' ')

print('The wrong way:')
display(thewrongway)
print(' ')
```

```python
# repeat with chain rule
gx = x**2 + 4*x**3
fx = gx ** 5

print('The function: ')
display(fx)
print(' ')

print('Its derivative: ')
display(sym.diff(fx))
```
