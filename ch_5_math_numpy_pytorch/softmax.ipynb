```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# manual calculation in numpy
def softmax(nums: list):

    nums_exp = np.exp(nums)
    denominator = np.sum(nums_exp)
    sigma = nums_exp / denominator

    return sigma

z = [1,2,3]
z = np.random.randint(-5, high=15, size=25)
print(z)
print(softmax(z))
print(sum(softmax(z)))
```

```python


fig, ax = plt.subplots(1, 2, figsize=(13, 7))

ax[0].plot(z, softmax(z), 'ko')
ax[0].set_title('$\sum\sigma$ = %g' % np.sum(softmax(z)))
ax[0].set_xlabel('Original number (z)')
ax[0].set_ylabel('Softmax $\sigma$')

ax[1].plot(z, softmax(z), 'ko')
ax[1].set_title('$\sum\sigma$ = %g' % np.sum(softmax(z)))
ax[1].set_xlabel('Original number (z)')
ax[1].set_ylabel('Softmax $\sigma$')
ax[1].set_yscale('log')

plt.show()
```

## __Pytorch__

```python
# creating an instance of nn.Softmax activation class
softmax_t = nn.Softmax(dim=0)

sigmas = softmax_t( torch.Tensor(z) )

print(sigmas)
print(sum(sigmas))
```

```python
plt.plot(z, sigmas, 'ko')

plt.plot(z, softmax(z), 'ko')
plt.title('The two methods correlate at r = %g' % np.corrcoef(sigmas, softmax(z))[0,1])
plt.xlabel('Original number (z)')
plt.ylabel('Softmax $\sigma$')
plt.show()
```
