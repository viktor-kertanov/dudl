```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n_1 = 30 # samples in dataset 1
n_2 = 40 # samples in dataset 2

mu_1 = 1
mu_2 = 1.1

# generating the data
data_1 = mu_1 + np.random.randn(n_1)
data_2 = mu_2 + np.random.randn(n_2)

# plot the data
plt.plot(np.zeros(n_1), data_1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n_2), data_2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=['Group 1', 'Group 2'])
plt.show()
```

```python
# t-test via stats package
# _ind = independent samples

t, p = stats.ttest_ind(data_2, data_1) # t-value, p-value
print(t)
print(p)

# The results show that 
```

```python
# common way to show t-test results in a plot
# p-value in a t-test should be less than 0.05 to be statistically significant

fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size': 12})

plt.plot(0 + np.random.randn(n_1) / 15, data_1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1 + np.random.randn(n_2) / 15, data_2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=['Group 1', 'Group 2'])

plt.title(f't = {t:.2f}, p={p:.5f}')

plt.show()
```
