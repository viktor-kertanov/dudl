```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import sympy as sym
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
x = sym.symbols('x')
fx = sym.cos(2*sym.pi*x) + x**2

deriv = sym.diff(fx)

display(fx)
display(deriv)
```

```python
def fx_val(x):
    return x**2 + np.cos(2*np.pi*x)

def deriv_val(x):
    return 2*x - 2*np.pi*np.sin(2*np.pi*x)

x_space = np.linspace(-2, 2, 2001)

fx_vals = fx_val(x_space)
deriv_vals = deriv_val(x_space)
```

```python
plt.plot(x_space, fx_vals, x_space, deriv_vals)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)', 'df'])
plt.show()
```

```python
localmin = np.random.choice(x_space, 1)
# localmin = 0

learning_rate = 0.01
training_epochs = 1000

for i in range(training_epochs):
    grad = deriv_val(localmin)
    localmin -= grad * learning_rate

plt.plot(x_space, fx_vals, x_space, deriv_vals)
plt.plot(localmin, fx_val(localmin), 'ro')
plt.plot(localmin, deriv_val(localmin), 'ro')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)', 'df'])
plt.show()
```

```python
localmin = np.random.choice(x_space, 1)
# localmin = 0 # vanishing gradient
# In practice we start in a random location in a multidimenstional space - but we have zero chances
# to face this in practice

learning_rate = 0.01
training_epochs = 1000

plt.plot(x_space, fx_vals, x_space, deriv_vals)

for i in range(training_epochs):
    grad = deriv_val(localmin)
    localmin -= grad * learning_rate
    plt.plot(localmin, fx_val(localmin), 'bo', markersize=2)
    plt.plot(localmin, deriv_val(localmin), 'ro', markersize=2)

plt.plot(localmin, fx_val(localmin), 'go', markersize=3)
plt.plot(localmin, deriv_val(localmin), 'go', markersize=3)

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)', 'df'])
plt.gcf().set_size_inches((12, 6))
plt.title(f'Local min x: {float(localmin):.2f}. y: {float(fx_val(localmin)):.2f}')
plt.show()
```
